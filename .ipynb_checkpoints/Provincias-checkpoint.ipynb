```python
import mapclassify 
import numpy as np

np.random.seed(12345) # so we all get the same results!

# let's try 5 intervals
K=5
# same interval width, easy interpretation
ei5 = mapclassify.EqualInterval(dataidhmap['IDH'], k=K)
# same interval width based on standard deviation, easy - but not as the previous one, poor when high skewness
msd = mapclassify.StdMean(dataidhmap['IDH'])
# interval width varies, counts per interval are close, not easy to grasp, repeated values complicate cuts                                
q5=mapclassify.Quantiles(dataidhmap['IDH'],k=K)

# based on similarity, good for multimodal data 
mb5 = mapclassify.MaximumBreaks(dataidhmap['IDH'], k=K)
# based on similarity, good for skewed data
ht = mapclassify.HeadTailBreaks(dataidhmap['IDH']) # no K needed
# based on similarity, optimizer
fj5 = mapclassify.FisherJenks(dataidhmap['IDH'], k=K)
# based on similarity, optimizer
jc5 = mapclassify.JenksCaspall(dataidhmap['IDH'], k=K)
# based on similarity, optimizer
mp5 = mapclassify.MaxP(dataidhmap['IDH'], k=K)   
```

```python
#Tabla de frecuencias
ht
```

```python
# group label
ht.yb
```

```python
# labels and counts
np.unique(ht.yb,return_counts=True)
          
```

```python
# ht.yb into a pandas Series

pd.Series(ht.yb).value_counts()
```

```python
# these are the cuts, noincluye el minimo valor
ht.bins
```

```python
# completo con el menor 
HT_bins=list(ht.bins)
HT_bins.insert(0,dataidhmap.IDH.min())
```

```python
#graficoun histograma
sea.displot(dataidhmap.IDH, bins=HT_bins,color='lightblue')
plt.show()
```

```python
#para hallar clasificacion correcta, usamos el Absolute deviation around class median para comparar
class5 = q5, ei5,msd, ht, mb5, fj5, jc5, mp5
# Collect ADCM for each classifier
fits = np.array([ c.adcm for c in class5])
# Convert ADCM scores to a DataFrame
adcms = pd.DataFrame(fits)
# Add classifier names
adcms['classifier'] = [c.name for c in class5]
# Add column names to the ADCM
adcms.columns = ['ADCM', 'Classifier']
ax = sea.barplot(
    y='Classifier', x='ADCM', data=adcms, palette='Pastel1'
)
plt.show()
```

```python
#nos quedamos con las que tienen menores ACDM
dataidhmap['QU'] = q5.yb 
dataidhmap['FJ'] = fj5.yb
dataidhmap['JC'] = jc5.yb
```

```python
class5new = q5,fj5, jc5
pd.DataFrame(
    {c.name: c.counts for c in class5new},
    index=['Class-{}'.format(i) for i in range(5)]
)
```

```python
#Elijo 3 variables Esperanza de vida, procentaje de educacion secundaria completa y Años de educacion
selected_variables = ['Pct_EduSecCompleta',
                     'EspDeVida', 
                     'AnhosEdu']
dataidhmap[selected_variables].describe()
```

```python
#visualizamos
sea.boxplot(data=dataidhmap[selected_variables])
plt.show()
```

```python
dataidhmap[selected_variables].hist()
plt.show();
```

```python
#Vemos la relación que tienen
sea.pairplot(
    dataidhmap[selected_variables], kind="reg", diag_kind="kde"
)
plt.show();
```

```python
#todos tiene una relacion positiva aunque su correlacion no es tan fuerte, ahora vamos a re escalar las variables
columnsToScale=selected_variables

from sklearn.preprocessing import robust_scale as rs
from sklearn.preprocessing import power_transform as pt
from sklearn.preprocessing import quantile_transform as qt

rs_result=rs(dataidhmap[columnsToScale])
pt_result=pt(dataidhmap[columnsToScale])
qt_result=qt(dataidhmap[columnsToScale], n_quantiles=100)
```

```python
df = pd.DataFrame(rs_result)
plt.boxplot(df.values, vert=False)
plt.show()
```

```python
df2 = pd.DataFrame(pt_result)
plt.boxplot(df2.values, vert=False)
plt.show()
```

```python
df3 = pd.DataFrame(qt_result)
plt.boxplot(df3.values, vert=False)
plt.show()
```

```python
#me quedo con el de qt
# new names
selected_variables_t=[s+'_t' for s in selected_variables]

# add colunms
dataidhmap[selected_variables_t]=qt_result
```

```python
#hago un mapa para cada variable con el scheme de cuantiles (el mejor)
f, axs = plt.subplots( ncols=1, nrows=3,figsize=(10, 10))
axs = axs.flatten()
# Start a loop over all the variables of interest
for i, col in enumerate(selected_variables_t):
    # select the axis where the map will go
    ax = axs[i]
    # Plot the map
    dataidhmap.plot(
        column=col,
        ax=ax,
        scheme="Quantiles",
        linewidth=0,
        cmap="RdPu",
    )
    # Remove axis clutter
    ax.set_axis_off()
    # Set the axis title to the name of variable being plotted
    ax.set_title(col)
# Display the figure
plt.show()
```

```python
# update
dataidhmap.to_file(os.path.join('maps',"dataMapPeru.gpkg"), layer='provincias', driver="GPKG")
```

```python
from scipy.cluster import hierarchy as hc


Z = hc.linkage(qt_result, 'ward')
# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('cases')
plt.ylabel('distance')
hc.dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=1,  # font size for the x axis labels
)
plt.show()
```

```python
#Veo como es un dendograma de 3 clusters
from sklearn.cluster import AgglomerativeClustering as agnes
# Set seed for reproducibility
np.random.seed(12345)
# Initialize the algorithm, requesting 3 clusters
model = agnes(linkage="ward", n_clusters=3).fit(dataidhmap[selected_variables_t])
# Assign labels to main data table
dataidhmap["hc_3"] = model.labels_
```

```python
#Creo otro con 5 para comparar
from sklearn.cluster import AgglomerativeClustering as agnes
# Set seed for reproducibility
np.random.seed(12345)
# Initialize the algorithm, requesting 3 clusters
model = agnes(linkage="ward", n_clusters=5).fit(dataidhmap[selected_variables_t])
# Assign labels to main data table
dataidhmap["hc_5"] = model.labels_
```
