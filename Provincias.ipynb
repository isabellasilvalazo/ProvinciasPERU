```python
#elijo alternativa A, consigo mapa y tabla de variables sobre las provincias de Peru
#importamos la tabla de data
import pandas as pd
import os

dataidh=pd.read_excel(os.path.join('data','IDH-y-Componentes-2003-2017.xlsx'),
                     dtype={'Ubigeo': object})
dataidh.head()
```

```python
#ahora importo mapa (geodataframe)
os.environ['USE_PYGEOS'] = '0'

import geopandas as gpd

dataidhmap=gpd.read_file(os.path.join('maps','PROVINCIAS_inei_geogpsperu_suyopomalia.zip'))

dataidhmap.head()
```

```python
#Notamos que podemos limpiar y preprocesar, en especial la data del idh
dataidh
```

```python
dataidh = dataidh.drop([0, 1, 208, 207, 206], axis=0).reset_index(drop=True)
dataidh
```

```python
#nos quedamos solo con los datos del año 2003 y quitamos columnas de datos nulos
dataidh=dataidh.iloc[:, [0, 1, 2,5, 13,21, 29, 37, 45]]
dataidh
```

```python
#Cambiamos nombres de las columnas
dataidh=dataidh.rename(columns={'Unnamed: 0': 'Ubigeo', 'Unnamed: 1': 'Departamento', 'Unnamed: 2': 'Provincia', 'Unnamed: 5': 'Habitantes', 'Unnamed: 13': 'IDH', 'Unnamed: 21': 'EspDeVida', 'Unnamed: 29': 'Pct_EduSecCompleta', 'Unnamed: 37': 'AnhosEdu', 'Unnamed: 45': 'IngresoPerCap'})
dataidh
```

```python
#Borramos las primeras filas
dataidh = dataidh.drop([0, 1, 2, 3, 4], axis=0).reset_index(drop=True)
dataidh
```

```python
dataidh = dataidh.drop([0, 1], axis=0).reset_index(drop=True)
dataidh
```

```python
#del geodataframe forrmaos algunas columnas
dataidhmap=dataidhmap.iloc[:, [0, 1,5,79]]
#cambio nombre de columnas
dataidhmap=dataidhmap.rename(columns={'NOMBDEP': 'DEPARTAMENTO', 'NOMBPROV': 'PROVINCIA'})
dataidhmap
```

```python
#Ahora comenzamos con el preprocesamiento
capitalizeColumns=lambda x: x.str.upper().str.strip()
dataidh[['Provincia','Departamento']]=dataidh[['Provincia','Departamento']].apply(capitalizeColumns)
dataidhmap[['PROVINCIA','DEPARTAMENTO']]=dataidhmap[['PROVINCIA','DEPARTAMENTO']].apply(capitalizeColumns)
```

```python
dataidh.info()
```

```python
dataidhmap.info()
```

```python
#Convierto numeros numericos de dataidh a float
dataidh['Habitantes'] = dataidh['Habitantes'].astype(float)
dataidh['IDH'] = dataidh['IDH'].astype(float)
dataidh['Pct_EduSecCompleta'] = dataidh['Pct_EduSecCompleta'].astype(float)
dataidh['EspDeVida'] = dataidh['EspDeVida'].astype(float)
dataidh['AnhosEdu'] = dataidh['AnhosEdu'].astype(float)
dataidh['IngresoPerCap'] = dataidh['IngresoPerCap'].astype(float)
```

```python
len(dataidh.Provincia),len(dataidhmap.PROVINCIA)
```

```python
#notamos que hay uno extra por lima
dataidh = dataidh.drop(dataidh.loc[dataidh['Provincia'] == 'LIMA'].index)
```

```python
len(dataidh.Provincia),len(dataidhmap.PROVINCIA)
```

```python
dataidh = dataidh.reset_index(drop=True)
```

```python
dataidh.Provincia.duplicated().sum(),dataidhmap.PROVINCIA.duplicated().sum()
```

```python
dataidh['Provincia'] = dataidh['Provincia'].astype(str)
dataidh['Departamento'] = dataidh['Departamento'].astype(str)
dataidhmap['PROVINCIA'] = dataidhmap['PROVINCIA'].astype(str)
dataidhmap['DEPARTAMENTO'] = dataidhmap['DEPARTAMENTO'].astype(str)
import unidecode

byePunctuation=lambda x: unidecode.unidecode(x)
dataidh[['Provincia','Departamento']]=dataidh[['Provincia','Departamento']].applymap(byePunctuation)
dataidhmap[['PROVINCIA','DEPARTAMENTO']]=dataidhmap[['PROVINCIA','DEPARTAMENTO']].applymap(byePunctuation)

```
